```python
import numpy as np
```

```python
x = np.array([x for x in range(100)])
x
```

```python
y = x+x
y
```

```python
plt.plot(x,y) # plota um grafico simples colocando o primeiro argumento como o eixo das abcissas
```

```python
plt.plot(x,y, color='red') #color = cor da linha (pode usar só uma letra mas isso gera ambiguidade como gray ou green)
plt.xlabel('KKKKKKKKKKK')#nome do eixo das abcissas
plt.ylabel('MMMMMMMMMMMMMM')#nome do eixo das ordenadas
plt.title('TITULO FODA 96') # titulo do grafico
```

```python
plt.plot(x,y, 'b--')# desse jeito além de ser azul (blue) é tracejada mas não funciona se escrever a cor completa ou usar o color='b--''
```

```python
plt.subplot(1, 3, 2)# garante multiplos plots
# o primeiro arg diz quantas linhas de graficos voce vai usar (na pratica quanto maior for menor vai ser a imagem)
# o  segundo arg diz quantas colunas de graficos você vai usar (diz quantos graficos serão plotados)
# o terceiro arg diz em qual das colunas esse grafico será plotado
plt.plot(x,y,'r--')
plt.subplot(3, 3, 1)
plt.plot(y,x, 'black')
plt.subplot(1, 3, 3)
plt.plot(x,-y,'g*-')
```

```python
# a melhor maneira de se criar graficos e instanciando classes de plt para usar os parametros dessas classes
```

```python
fig = plt.figure() #objeto usado para criar graficos que se assemelha a uma tela em branco
eixo = fig.add_axes([0.1,0.1,0.9,0.9]) #adiciona um eixo onde vão ser colocados os graficos 
# primeiro arg define o quanto a direita o grafico está (nesse caso é 10% a direita)
# segundo arg define o quanto a direita o grafico está (nesse caso é 10% acima)
# terceiro arg define a largura (nesse caso é 90% da largura da figura 'fig')
# quarto arg define a altura ((nesse caso é 90% da altura da figura 'fig'))
# obs 0.1 = 10%, 4.5 = 450%

eixo.plot(x,y)
eixo.set_xlabel('KKKKK')#pra o grafico aparecer vc só precisa de um desses tres sets
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
#pra o grafico ser mostrado a fig precisa ser instanciada no mesmo bloco que os demais comandos
```

```python
fig = plt.figure()
eixo = fig.add_axes([0.1,0.1,3.5,0.9]) # um grafico com maior largura
eixo.plot(x,y)
eixo.set_xlabel('KKKKK')#pra o grafico aparecer vc só precisa de um desses tres sets
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
```

```python
fig = plt.figure()
eixo = fig.add_axes([0.1,0.1,0.9,3.5]) # um grafico com maior altura
eixo.plot(x,y)
eixo.set_xlabel('KKKKK')#pra o grafico aparecer vc só precisa de um desses tres sets
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
```

```python
# os 2 primeiros argumentos de aces só ficam fazem diferença se você poltar vairos objetos de uma vez
fig = plt.figure()
eixo = fig.add_axes([99,99,0.9,0.9]) # mesmo estando mais a direita e mais para cima em relaão a origem as mudanças no eixo não são visiveis
eixo.plot(x,y)
eixo.set_xlabel('KKKKK')#pra o grafico aparecer vc só precisa de um desses tres sets
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
```

```python
fig = plt.figure()
eixo = fig.add_axes([0.1,0.1,0.9,0.9])
eixo2 = fig.add_axes([0.4,0.1,0.4,0.4])# usando um grafico menor para mostrar como os 2 primeiros parametros interferem
#nota-se que o grafico menor está mais a direita pois o primeiro arg é maior
eixo.plot(x,y, 'g')
eixo.set_xlabel('KKKKK')
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
eixo2.plot(y,x,'b')
eixo2.set_xlabel('ebcissa do segundo grafico')
eixo2.set_ylabel('ordenada do segundo grafico')
eixo2.set_title('TITULO FODA 666')
```

```python
fig = plt.figure()
eixo = fig.add_axes([0.1,0.1,0.9,0.9])
eixo2 = fig.add_axes([0.1,0.4,0.4,0.4])# usando um grafico menor para mostrar como os 2 primeiros parametros interferem
#nota-se que o grafico menor está mais a cima pois o cima arg é maior
eixo.plot(x,y, 'g')
eixo.set_xlabel('KKKKK')
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
eixo2.plot(y,x,'b')
eixo2.set_xlabel('ebcissa do segundo grafico')
eixo2.set_ylabel('ordenada do segundo grafico')
eixo2.set_title('TITULO FODA 666')
```

```python
fig = plt.figure()
eixo = fig.add_axes([0.1,0.1,0.9,0.9])
eixo2 = fig.add_axes([0.7,0.4,0.4,0.4])# usando um grafico menor para mostrar como os 2 primeiros parametros interferem
#nota-se que o grafico menor está mais a cima pois o segundo arg é maior  e mais a direita pq o primeiro arg é maior
eixo.plot(x,y, 'g')
eixo.set_xlabel('KKKKK')
eixo.set_ylabel('MMMMMMMMMMMMMMMMMMMM')
eixo.set_title('TITULO FODA 99')
eixo2.plot(y,x,'b')
eixo2.set_xlabel('ebcissa do segundo grafico')
eixo2.set_ylabel('ordenada do segundo grafico')
eixo2.set_title('TITULO FODA 666')
```
